## CONVERT TO GREYSCALE

```python
image_gray = cv2.cvtColor(image_color, cv2.COLOR_BGR2GRAY)
plt.imshow(image_gray, cmap = 'gray')
```

```python
image_gray.shape
```

```python
# Save the image in greyscale
# cv2.imwrite("antilia_in_grey.jpg", image_gray)
```

```python
# find corners in an image 

threshold_1 = 20
threshold_2 = 250

image_canny = cv2.Canny(image_gray, threshold_1, threshold_2)

cv2.imshow('Canny', image_canny)
cv2.waitKey()
cv2.destroyAllWindows()
```

# it's time to detection!, FEATURE DETECTION

```python
# Create FAST Detector object
fast = cv2.FastFeatureDetector_create()

keypoints = fast.detect(image_gray, None)

print ("Number of keypoints Detected: ", len(keypoints))

# draw key points was not working as on 6-mar-2019 so changed it as below
# image = cv2.drawKeypoints(image_color, keypoints, None, color = (255,0,0))
#image = cv2.drawKeypoints(image_color, keypoints, None)

# added this part as drawkeypoint is not working
img2 = image_gray.copy()
for marker in keypoints:
    img2 = cv2.drawMarker(img2, tuple(int(i) for i in marker.pt), color=(0, 255, 0))
plt.imshow(img2)
plt.show()
# _______________________________________

cv2.imshow('Feature Method - FAST', img2)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# Create ORB object, we can specify the number of key points we desire
orb = cv2.ORB_create()

# Determine key points
keypoints = orb.detect(image_gray, None)

# Obtain the descriptors
keypoints, descriptors = orb.compute(image_gray, keypoints)
print("Number of keypoints Detected: ", len(keypoints))

# Draw rich keypoints on input image
# image = cv2.drawKeypoints(image_color, keypoints,None, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

# added this part as drawkeypoint is not working
img2 = image_gray.copy()
for marker in keypoints:
    img2 = cv2.drawMarker(img2, tuple(int(i) for i in marker.pt), color=(0, 255, 0))
plt.imshow(img2)
plt.show()
# _______________________________________

cv2.imshow('Feature Method - ORB', img2)
cv2.waitKey()
cv2.destroyAllWindows()
```

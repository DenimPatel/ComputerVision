## IMAGE PYRAMIDING (Gaussian)

# LOAD/DISPLAY IMAGE

```python
import cv2
import numpy as np
import matplotlib.image as mpimg
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
image_color = cv2.imread('antilia.jpg')
# image_color = cv2.imread('truck_back_color.jpg')

cv2.imshow('Original Image', image_color)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
plt.imshow(image_color)
```

```python
image_color.shape
```

# PERFROM PYRAMIDING (DOWN)

```python
scaled_down_1 = cv2.pyrDown(image_color)
print(scaled_down_1.shape)
plt.imshow(scaled_down_1)
```

```python
scaled_down_2 = cv2.pyrDown(scaled_down_1)
print(scaled_down_2.shape)
plt.imshow(scaled_down_2)
```

```python
scaled_down_3 = cv2.pyrDown(scaled_down_2)
print(scaled_down_3.shape)
plt.imshow(scaled_down_3)
```

```python
cv2.imshow('Original', image_color)
cv2.imshow('Smaller #1', scaled_down_1)
cv2.imshow('Smaller #2', scaled_down_2)
cv2.imshow('Smaller #3', scaled_down_3)

cv2.waitKey()
cv2.destroyAllWindows()
```

# PERFROM PYRAMIDING (UP)

```python
scaled_up = cv2.pyrUp(image_color)
print(scaled_up.shape)
plt.imshow(scaled_up)
```

```python
cv2.imshow('Larger ', scaled_up)
cv2.waitKey()
cv2.destroyAllWindows()
```

## HISTOGRAM OF ORIENTED GRADIENTS

# LOAD/DISPLAY IMAGE

```python
import cv2
import numpy as np
import matplotlib.image as mpimg
from matplotlib import pyplot as plt
from skimage.feature import hog # The Histogram of Oriented Gradient (HOG)

%matplotlib inline
```

```python
image_color = cv2.imread('antilia.jpg')
cv2.imshow("someone's house", image_color)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
plt.imshow(image_color)
```

## CONVERT TO GREYSCALE

```python
image_gray = cv2.cvtColor(image_color, cv2.COLOR_BGR2GRAY)
plt.imshow(image_gray, cmap = 'gray')
```

```python
image_gray.shape
```

```python
# Sobel Calculations
x_sobel = cv2.Sobel(image_gray, cv2.CV_64F, 0, 1, ksize = 7)
cv2.imshow('Sobel - X direction', x_sobel)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# Sobel Calculations
y_sobel = cv2.Sobel(image_gray, cv2.CV_64F, 1, 0, ksize = 7)
cv2.imshow('Sobel - Y direction', y_sobel)
cv2.waitKey()
cv2.destroyAllWindows()
```

# PERFORM HISTOGRAM OF ORIENTED GRADIENT

```python
features, hog_image = hog(image_gray, orientations = 9,
                          pixels_per_cell = (16, 16),cells_per_block = (1, 1),
                          transform_sqrt = False,visualise = True, feature_vector = False)

features.shape

```

```python
features[10,11:]
```

```python
print(hog_image.shape)
plt.imshow(hog_image)
```

```python
# Return image after stretching or shrinking its intensity levels.
from skimage import exposure

hog_image_rescaled = exposure.rescale_intensity(hog_image, in_range=(0, 2))
```

```python
plt.figure()    
plt.imshow(hog_image_rescaled, cmap = 'gray')
plt.title('HOG Features')
    
```

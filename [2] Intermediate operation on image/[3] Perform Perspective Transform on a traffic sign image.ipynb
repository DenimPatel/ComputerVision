## PERFORM PERSPECTIVE TRANSFORM

```python
import cv2
import numpy as np
import matplotlib.image as mpimg
from matplotlib import pyplot as plt
%matplotlib inline

image = cv2.imread('Speed_Sign.jpg')
cv2.imshow('Original Image', image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
plt.imshow(image)
```

```python
height, width = image.shape[:2]
print(height)
print(width)
```

## we want to perform perspactive transform of the sign 

for that we need to provide 4 points for reference
- 4 points in origional image
- 4 points in desired position

## we need to specify the points in the following order 
- First entry: top-left
- Second entry: top-right
- Third: bottom-right
- Fourth: bottom-left

```python
# Cordinates of the 4 corners of the original image
Source_points = np.float32([[200,60], [450,150], [520, 500], [170,470] ])

# Cordinates of the 4 corners of the desired output
Destination_points = np.float32([[0,0], [width,0], [width,height], [0,height]])
```

```python
# Use the two sets of four points to compute 
# the Perspective Transformation matrix, M    
M = cv2.getPerspectiveTransform(Source_points, Destination_points)
 
warped = cv2.warpPerspective(image, M, (width, height))
 
cv2.imshow('warped Image', warped)
cv2.waitKey()
cv2.destroyAllWindows()
```

# ROTATION, TRANSLATION AND RESIZING


# LOAD/DISPLAY IMAGE

```python
import cv2
import numpy as np
import matplotlib.image as mpimg
from matplotlib import pyplot as plt
%matplotlib inline

image = cv2.imread('road.jpg')
cv2.imshow('Original Image', image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
print(image.shape)
height, width = image.shape[:2]
print(height)
print(width)
```

# PERFORM ROTATIONS

```python
M_rotation = cv2.getRotationMatrix2D((width/2, height/2), 90, 0.4) # Rotate around the center

# this is really a great tool for image warping!
rotated_image = cv2.warpAffine(image, M_rotation, (width, height)) # insert size of the output image

cv2.imshow('Rotated_road!', rotated_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
print(M_rotation)
# same output as above cell just to see it in notebook
plt.imshow(rotated_image)
```

```python
warped_image = cv2.warpAffine(image, M_rotation, (width*2, height*2)) # insert size of the output image
plt.imshow(warped_image)
```

## # now understand that we need any operation scaling, translation, rotation, or any linear operation on image using warp affine

## we just need to pass image, transformation matrix, image size 

# PERFORM TRANSLATIONS

```python
Translational_Matrix = np.float32([[1, 0, 200], 
                                  [0, 1, 100]])
```

```python
translated_image = cv2.warpAffine(image, Translational_Matrix, (width, height))
cv2.imshow('Translated Image', translated_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# same output as above
plt.imshow(translated_image)
```

# PERFORM RESIZING


## we can increase or decrease the resolution by changing fx and fy

```python
resized_image = cv2.resize(image, None, fx=2, fy=2, interpolation = cv2.INTER_CUBIC) # Try 0.5
cv2.imshow('Resized Image', resized_image) 
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# same output as above
plt.imshow(resized_image)
```

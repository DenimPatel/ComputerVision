## HISTOGRAM OF COLOR

# LOAD/DISPLAY IMAGE

```python
import cv2
import numpy as np
import matplotlib.image as mpimg
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
image_color = cv2.imread('antilia.jpg')

# image_color = cv2.imread('red_night.jpg')
# image_color = cv2.imread('red_car_2.jpg')

cv2.imshow('Original Image', image_color)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
plt.imshow(image_color)
```

```python
image_color.shape
```

```python
color = ('b', 'g', 'r')

for i, col in enumerate(color):
    histogram2 = cv2.calcHist([image_color], [i], None, [256], [0, 256])
    plt.plot(histogram2, color = col)
    plt.xlim([0,256])
    
plt.show()
```

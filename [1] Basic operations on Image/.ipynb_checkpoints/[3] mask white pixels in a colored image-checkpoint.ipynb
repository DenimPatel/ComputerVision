#  SELECT WHITE PIXELS OUT OF THE IMAGE

# Import all the necessary libraries

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

# Read the image and display it

```python
image_color = mpimg.imread('road.jpg')
image_color.shape
```

# Let's sort out the white colour out of the image

```python
image_copy = np.copy(image_color)
image_copy.shape
```

```python
# As we have 3 colour buckets R(red), G(Green) and B(Blue) 
# we need to make threshold for each color bin to detect any pixel

image_copy[ (image_copy[:,:,0] < 100) | (image_copy[:,:,1] < 100) | (image_copy[:,:,2] < 100) ] = 0 
# any value that is not white colour put 0 there
```

```python
# Display the image                 
plt.imshow(image_copy, cmap = 'gray')
plt.show()
```

#  Edge Detection using Sobel/Laplacian and Canny

# Import all the necessary libraries

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np

import cv2
```

# Read the image and display it using OpenCV

```python
import cv2 

image = cv2.imread('road.jpg')

cv2.imshow('Road that never ends!', image)
cv2.waitKey()
cv2.destroyAllWindows()
```

## print out the dimensions of convert to grayscale image

```python
image.shape
```

```python
gray_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imshow('road_in_gray!', gray_img)
cv2.waitKey()
cv2.destroyAllWindows()
```

# Let's Get Sobel X and Y

## Sobel x gives the horizontal edges detected and Sobel y gives vertical edges

## Sobel's can be implemented using Convolution easily!

```python
# Sobel Calculations
x_sobel = cv2.Sobel(gray_img, cv2.CV_64F, 0, 1, ksize = 7)
cv2.imshow('Sobel - X direction', x_sobel)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# same output as above cell just to see it in notebook
plt.imshow(x_sobel)
```

```python
y_sobel = cv2.Sobel(gray_img, cv2.CV_64F, 1, 0, ksize = 7)
cv2.imshow('Sobel - Y direction', y_sobel)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# same output as above cell just to see it in notebook
plt.imshow(y_sobel)
```

# Laplacian 

## The Laplacian of an image highlights regions of rapid intensity change and is therefore often used for edge detection 

```python
laplacian = cv2.Laplacian(gray_img, cv2.CV_64F)
cv2.imshow('Laplacian', laplacian)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# same output as above cell just to see it in notebook
plt.imshow(laplacian)
```

# Canny 

```python
# threshold1  first threshold for the hysteresis procedure.
# threshold2  second threshold for the hysteresis procedure.# Let's get Canny Edge

threshold_1 = 120
threshold_2 = 200


canny = cv2.Canny(gray_img, threshold_1, threshold_2)

cv2.imshow('Canny', canny)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# same output as above cell just to see it in notebook
plt.imshow(canny)
```

# CODE TO CONVERT COLOR IMAGE TO GRAYSCALE

# Import all the necessary libraries

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

# Read/Display an image

```python
image_color = mpimg.imread('road.jpg')
```

```python
plt.imshow(image_color)
```

```python
# to check the size of the image
image_color.shape
```

# Convert an image into Grayscale

```python
image_gray = cv2.cvtColor(image_color, cv2.COLOR_BGR2GRAY)
plt.imshow(image_gray, cmap = 'gray')
```

```python
image_gray.shape
```

```python
image_gray
```

```python
np.set_printoptions(threshold=np.nan)
image_gray
```

# Save the Grayscale image

```python
# Save the image in greyscale
cv2.imwrite("road_grayscaled.jpg", image_gray)
```

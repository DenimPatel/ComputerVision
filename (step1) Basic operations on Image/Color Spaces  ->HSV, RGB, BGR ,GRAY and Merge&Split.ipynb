# Let's explore some pixels and convert to gray color

```python
# to access RGB values of any specific pixel
# note that OpenCV stores values in BGR sequence!
B, G, R = image[220, 300] 
print(B, G, R)
```

# Let's convert from RGB to HSV

```python
image = cv2.imread('road.jpg')
cv2.imshow('Road_endless_road!', image)

hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
cv2.imshow('HSV_road_Image', hsv_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# same output as above cell just to see it in notebook
plt.imshow(hsv_image)
```

## visualize Hue, Saturation and Value 

```python
plt.imshow(hsv_image[:, :, 0])
plt.title('Hue channel')
```

```python
plt.imshow(hsv_image[:, :, 1])
plt.title('Saturation channel')
```

```python
plt.imshow(hsv_image[:, :, 2])
# add a title to the image
plt.title('Value channel')
```

# split and merge R, G, B channels 

```python
image = cv2.imread('road.jpg')
B, G, R = cv2.split(image)
B.shape
```

```python
R.shape
```

```python
# Let's view the R, B, G channels
cv2.imshow("Blue Channel!", R) 
cv2.waitKey()
cv2.destroyAllWindows()

# We didn't expect these Results!
```

```python
# same output as above cell just to see it in notebook
plt.imshow(R)
```

## in order to split color channels from the image, we need to do something as below!

```python
# Let's try to create our own 3D image out of the blue channel!
zeros = np.zeros(image.shape[:2], dtype = "uint8")
blue_image = cv2.merge([B, zeros, zeros])
cv2.imshow("Blue Channel!", blue_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# same output as above cell just to see it in notebook
plt.imshow(blue_image)
# it's just showing red becuse of inverted colors
```

```python
# Let's merge RGB channels to create our original image
image_merged = cv2.merge([B, G, R]) 
cv2.imshow("Merged Image!", image_merged) 

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# same output as above cell just to see it in notebook
plt.imshow(image_merged)
```

```python
# Let's merge our original image while adding more green! 
image_merged_added = cv2.merge([B, G+100, R]) 
cv2.imshow("Merged Image!", image_merged) 

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# same output as above cell just to see it in notebook
plt.imshow(image_merged_added)
```

#  Convolutions - Blurring and Sharpening

# Import all the necessary libraries

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

# Read the image and display it using OpenCV

```python
import cv2 

image = cv2.imread('road.jpg')

cv2.imshow('roads_never_ends!', image)

# 'waitKey' allows us to input information when a image window is open
# By leaving it blank it just waits for anykey to be pressed before continuing. 
cv2.waitKey()
cv2.destroyAllWindows()
```

## Let's print out the dimensions of the image

```python
image.shape
```

```python
# convert image to gray
gray_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imshow('gray_road!', gray_img)
cv2.waitKey()
cv2.destroyAllWindows()
```

# Convolution Convolution Convolution

# Let's apply sharpening Kernel

```python
# Sharpening Kernel #1
Sharp_Kernel = np.array([[0,-1,0], 
                        [-1,5,-1], 
                        [0,-1,0]])

# this kernel only uses 4 connected cells 

# Sharpening Kernel #2
#Sharp_Kernel = np.array([[-1,-1,-1], 
#                         [-1,9,-1], 
#                         [-1,-1,-1]])

# this kernel uses 8 connected cells

Sharpened_Image = cv2.filter2D(gray_img, -1, Sharp_Kernel)

# Notice that the kernels mentioned above are normalized, means its sum will be always 1 
# otherwise the intensity value of the images will be amplified
```

```python
cv2.imshow('Sharpened Image', Sharpened_Image)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# same output as above cell just to see it in notebook
plt.imshow(Sharpened_Image)
```

# Blurring Kernel

```python
# Blurring Kernel 
Blurr_Kernel = np.ones((3,3))
Blurr_Kernel
# you may have noticed that the kernel is not normalized!!
```

```python
Blurred_Image = cv2.filter2D(gray_img, -1, Blurr_Kernel)
```

```python
cv2.imshow('Blurred Image', Blurred_Image)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# same output as above cell just to see it in notebook
plt.imshow(Blurred_Image)

# Yes!! the output is as expected
```

# Task: normalize Kernel!

```python
# Blurring Kernel normalized 
Blurr_Kernel = np.ones((3,3)) * 1/9   #did you understand why we divided by 9??

Blurred_Image = cv2.filter2D(gray_img, -1, Blurr_Kernel)
cv2.imshow('Blurred Image', Blurred_Image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# same output as above cell just to see it in notebook
plt.imshow(Blurred_Image)
```

# Larger (more powerful) kernel size

```python
# kernel selected as 8
Blurr_Kernel = np.ones((8,8))
Blurr_Kernel
```

```python
# Blurring Kernel needs to be normalized 
Blurr_Kernel = np.ones((8,8)) * 1/64
Blurr_Kernel
```

```python
Blurred_Image = cv2.filter2D(gray_img, -1, Blurr_Kernel)

cv2.imshow('Blurred Image', Blurred_Image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# same output as above cell just to see it in notebook
plt.imshow(Blurred_Image)
```

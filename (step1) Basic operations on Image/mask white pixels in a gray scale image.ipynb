# CODE TO PERFORM COLOR SELECTION

# Import all the necessary libraries

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

# Read the image and display it

```python
image_color = mpimg.imread('road.jpg')
image_gray = cv2.cvtColor(image_color, cv2.COLOR_BGR2GRAY)
plt.imshow(image_gray, cmap = 'gray')
image_gray.shape
```

# Let's sort out the white color out of the image

```python
image_copy = np.copy(image_gray)
image_copy.shape
```

```python
# in this case we need to select the colour threshold value for sorting
image_copy[ (image_copy[:,:] < 100) ] = 1  # any value that is not white colour
```

```python
# Display the image                 
plt.imshow(image_copy, cmap = 'gray')
plt.show()
```

```python
# Save the image in greyscale
cv2.imwrite("road_white_masked.jpg", image_copy)
```
